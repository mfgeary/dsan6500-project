```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
res152_loss = pd.read_csv("./output/resnet152_epoch_losses.csv")
res101_loss = pd.read_csv("./output/resnet101_epoch_losses.csv")
vg11_loss = pd.read_csv("./output/losses_accuracies_vgg11_0001.csv")
vg16_loss = pd.read_csv("./output/losses_accuracies_vgg16_0001.csv")


max_epochs = max(len(res152_loss), len(res101_loss), len(vg11_loss), len(vg16_loss))

fig = plt.figure()
plt.plot(res101_loss.index + 1, res101_loss['Test Loss'], 'k--', label = 'ResNet 101', marker='D', markersize=3, linewidth=0.8)
plt.plot(res152_loss.index + 1, res152_loss['Test Loss'], 'k-', label = 'ResNet 152', marker='o', markersize=3, linewidth=0.8)
# plt.plot(vg11_loss.index, vg11_loss['test_loss'], 'k:', label = 'VGG11', marker='s', markersize=3, linewidth=0.8)
# plt.plot(vg16_loss.index, vg16_loss['test_loss'], 'k:', label = 'VGG16', marker='^', markersize=3, linewidth=0.8)

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Test Loss per Epoch')
plt.legend()
plt.savefig('./output/test_loss_comparison.jpg')
plt.show()
```

```python
fig = plt.figure()
plt.plot(res101_loss.index + 1, res101_loss['Train Loss'], 'k--', label = 'ResNet 101', marker='D', markersize=3, linewidth=0.8)
plt.plot(res152_loss.index + 1, res152_loss['Train Loss'], 'k-', label = 'ResNet 152', marker='o', markersize=3, linewidth=0.8)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Train Loss per Epoch')
plt.legend()
plt.savefig('./output/train_loss_comparison.jpg')
plt.show()
```
